```python
import matplotlib.pyplot as plt
import random
import numpy as np
import sys

%matplotlib inline
plt.style.use('default')
plt.rcParams['figure.figsize'] = [8,4]
plt.rcParams["axes.edgecolor"] = "black"
```

```python
# model parameters
a = 0.7
b = 0.5
c = 0.3
e = 0.2

#Time parameters
dt = 0.001
max_time = 100

# initial time and populations
t = 0
x = 1.0
y = 0.5

# init lists
t_list = [t]
x_list = [x]
y_list = [y]

for t in np.linspace(t, max_time, num = 100000):
    
    # calc new values for t, x, y
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Plot the results    
plt.plot(t_list, x_list, 'black',linewidth = 1, ls = '-')
plt.plot(t_list, y_list, 'gray',linewidth = 1, ls = '-')
plt.legend(['Prey','Predator'])
plt.title('Lotka-Volterra plot (Baseline)')
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Population (arbitrary units)')
```

```python
# model params
a = 0.7
b = 0.5
c = 0.3
e = 0.2

#time params
dt = 0.001
max_time = 100

# initial time and populations
t = 0; x = 0.5; y = 0.5

# empty lists in which to store time and populations
t_list = []; x_list = []; y_list = []

# initialize lists
t_list.append(t); x_list.append(x); y_list.append(y)

# simulating deaths
time_bomb_armed = True

for t in np.linspace(t, max_time, num = 100000):
    
    if time_bomb_armed and t>53.5:
        x = 0.1 * x
        y = 0.1 * y
        time_bomb_armed = False

    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    # store new values in lists
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Plot the results    
plt.plot(t_list, x_list, 'black',linewidth = 1, ls = '-')
plt.plot(t_list, y_list, 'gray',linewidth = 1, ls = '-')
plt.legend(['Prey','Predator'])
plt.title('Lotka-Volterra plot (Bombed)')
plt.xlabel('Time (arbitrary units)')
plt.ylabel('Population (arbitrary units)')
```
